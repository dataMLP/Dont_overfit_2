```python
import mglearn
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
X = train.drop(['id', 'target'], axis=1)
y = train['target']
test = test.drop(['id'], axis=1)
```

```python
test.head()
```

```python
''' 특성이 300개가 있습니다. 즉 고차원 모델.여기서 각 특성이 타겟값에 얼마나 큰 영향을 미치는지에 대해서 살펴보면 되겠죠'''
```

```python
''' 
1. 선형 모델
2. 트리(결정 트리, 랜덤포레스트, 그레이디언트 부스팅, 엑스트라 트리, 히스토그램 기반 부스팅)
3. 서포트 벡터 머신
'''
'''
MLP를 사용하지 않은 이유 : 가중치를 해석하기 매우 어렵기 때문
후술하지만 서포트 벡터 머신의 경우 성능은 가장 뛰어나지만
특성별 가중치를 구하는것이 불가능하다 -> 결국 의미 없는 모델

l1 제약 l2 제약 한 이유
랜덤 트리를 안하고 결정 트리를 하게 된 이유
딥러닝(신경망) 모델 채택 안한 이유
'''
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
plt.figure(figsize=(50, 30))
plt.boxplot(X_train)
plt.yscale("symlog")
plt.xlabel("특성 목록")
plt.ylabel("특성 크기")
```

데이터의 크기가 매우 비슷합니다.
보통 이러한 경우에는 SVM(서포트 백터 머신)의 성능이 굉장히 좋아집니다.
하지만 중요한 특성을 가려내는 입장에서 그다지 유용한 모델이 아닐 수 있습니다.

```python
for C in range(-3, 2):
    C = 10**C
    model = LogisticRegression(C = C, max_iter=50000).fit(X_train, y_train)
    print("C={:.3f} 인 릿지의 훈련 정확도: {:.2f}".format(
        C, model.score(X_train, y_train)))
    print("C={:.3f} 인 릿지의 테스트 정확도: {:.2f}".format(
        C, model.score(X_test, y_test)))
    print("사용한 특성의 개수:", np.sum(model.coef_ != 0))
```

```python
plt.figure(figsize=(50, 30))
for C, marker in zip([0.01, 0.1, 1], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(solver='liblinear', C=C, penalty="l1", max_iter=50000).fit(X_train, y_train)
    print("C={:.2f} 인 l1 로지스틱 회귀의 훈련 정확도: {:.2f}".format(
        C, lr_l1.score(X_train, y_train)))
    print("C={:.2f} 인 l1 로지스틱 회귀의 테스트 정확도: {:.2f}".format(
        C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label="C={:.3f}".format(C))    
plt.xticks(range(0, 300))
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("특성")
plt.ylabel("계수 크기")

plt.ylim(-1, 1)
plt.legend(loc=3)
plt.savefig('savefig_default.png')
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X_train, y_train)
print("훈련 세트 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["1", "0"],
               impurity=False, filled=True)
```

```python
import graphviz

with open("tree.dot", encoding='UTF8') as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
def plot_feature_importances_cancer(model):
    plt.figure(figsize=(50, 30))
    n_features = 300
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features))
    plt.xlabel("특성 중요도")
    plt.ylabel("특성")
    plt.ylim(-1, n_features)
    plt.savefig('plot_tree.png')
    print(model.feature_importances_)
plot_feature_importances_cancer(tree)
```
